**Notebook #1-Data Cleaning and EDA**

I started out by importing the necessary libraries and taking a look at the data. The process included examining each column to check for data consistency, missing values and outliers. Missing values in the waterfront and year renovated columns became 0. "?" in the view column became 0. In the bedroom column, the value 33 seems to be a typo. Given that it has 2 bathrooms and the price is 640000, it should say 3 bedrooms. Bedrooms over 7 got dropped, because the data seemed unrealiable and there were so few entries.

**Feature Engineering**  I renamed the yr_renovated column to renovated. The dataset is from 2015, so I used the year built column and created a new column called "age" by calculating the difference between 2015 and the year that the house was built. For the renovated column, if the house is renovated, the value was changed to 1 (instead of the year it was renovated), and if the value was 0, nan or ? the value became 0. Additionally, if the house was less than 5 years old, it was considered renovated. A new column was created that calculated the distance from downtown Seattle using the longitude and latitude coordinates.The bathrooms and floors columns got rounded up to the nearest whole number, and the data types were changed to integers.

```python
df.rename(columns={'yr_renovated': 'renovated'}, inplace=True)
```

```python
df['age'] = 2015 - df.yr_built
df = df.drop(columns=['yr_built'], axis=1)
```

```python
df['renovated'].value_counts()
```

```python
renovated = df.renovated.values
age = df.age.values
values = np.where(renovated <= 10, 1, 0)
df['renovated'] = np.where(age <= 5, 1, values)
```

```python
from geopy import distance

lat_long = df['lat'].astype(str) + ',' + df['long'].astype(str)
lat_long = list(map(eval, lat_long)) 

Seattle = (47.6062, -122.3321)
miles = []
for i in lat_long:
    miles.append(round(distance.distance(i, Seattle).miles, 1))
df['distance'] = miles
```

```python
df[['month', 'date','year']] = df['date'].str.split('/', expand = True)
```

```python
df['waterfront']=df['waterfront'].astype(int)
```

```python
df['floors']=df['floors'].apply(np.ceil)
df['bathrooms']=df['bathrooms'].apply(np.ceil)
df['floors']=df['floors'].astype(int)
df['bathrooms']=df['bathrooms'].astype(int)
```

**Dropping Columns** the following columns were dropped because updated versions of them were created and or they weren't relevent.
